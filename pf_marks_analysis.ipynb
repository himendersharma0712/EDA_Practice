```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

import warnings
warnings.filterwarnings('ignore')
```

```python
pf_result = pd.read_excel('Result.xlsx')
```

```python
pf_result.head()
```

```python
pf_result = pf_result.drop(columns=['Unnamed: 11','Unnamed: 12'],axis=1)
```

```python
pf_result.head()
```

```python
pf_result.groupby('Roll No')['Total'].sum().sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(12,5))
sns.histplot(pf_result['Total'],bins=10)
plt.xlabel('Total marks')
plt.show()
```

```python
sns.boxplot(data=pf_result['Total'])
plt.ylabel('Total Marks')
plt.show()
```

```python
pf_result['Total'].mean() # mean score is 54.82
```

```python
pf_result['Total'].median() # median score is 62
```

```python
pf_result['Total'].mode() # 69 is the most frequently appearing value
```

```python
sns.scatterplot(data=pf_result,x=pf_result['Att %'],y=pf_result['Total'])
plt.ylabel('Total marks')
plt.xlabel('Attendance %')
plt.show()
```

```python
sns.lineplot(data=pf_result,x=pf_result['Att %'],y=pf_result['Total'])
plt.ylabel('Total marks')
plt.xlabel('Attendance %')
plt.show()
```
