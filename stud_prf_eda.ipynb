# Outlier detection

*To detect>> distribution,boxplot and violinplot*

```python
students_grades_df = df.copy()
students_grades_df
```

```python
students_grades_df = students_grades_df.drop(['G1','G2','G3'],axis=1)
```

```python
students_grades_df
```

```python
#Lower fence = Q1 - 1.5*IQR
#Upper fence = Q3 + 1.5*IQR
```

```python


def detect_outliers(columns):
    
    outlier_indices = []
    
    for col in columns:
        Q1 = students_grades_df[col].quantile(0.25)
        Q3 = students_grades_df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # mask is a bool condition
        mask = (students_grades_df[col] >= (Q1 - 1.5*IQR)) & (students_grades_df[col] <= (Q3 + 1.5*IQR))
        mask = mask.to_numpy() # bool to numpy array
        
        # find the indices of values that do not satisfy the mask condition
        false_indices = np.argwhere(~mask)
        outlier_indices.append(false_indices)
        
    return np.unique(np.concatenate(outlier_indices).ravel())

# np.concatenate(outlier_indices) >> combine all the outlier index arrays into a single array
# so each column will have one array
# ravel -> flattens the multi dim array into 1D array
```

```python
num_cols = ['age','absences']
detect_outliers(num_cols)
```
